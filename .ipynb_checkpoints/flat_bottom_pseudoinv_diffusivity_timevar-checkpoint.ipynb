# Do the tracer inverse in 2D run. 

**Setup**

Almost the exact same setup as Abernathey et al 2013.  
Some slight modifications were done to the channel size, and winds. 
The channel is now 700km X 1200km. Depth of channel is 1500m.  
20 tracer experiments were carried out. Each experiment was an initial condition release of 8 tracers, aligned in zonally uniform but meridionally and vertically varying configurations. The configurations were linear or sinusoidal functions. Every year on day 0 the initial condition was let go, and the system was allowed to evolve for 1 year. The next year, the same experiment would be repeated again. 
The tracer equation was solved online and the tracer fluxes were evaluated online and saved every 2 months (6 outputs per year). 

## ** Every 2 month of simulation is refered to as 1 month in this notebook **

** Analysis **  
Similar to Abernathey et al 2013 we will invert the following equation  

\begin{equation}
\overline{\mathbf{u}'c'} = - \mathbf{R} \nabla C,
\end{equation}

where $\mathbf{u}'$ and $c'$ is the deviation from a mean of the velocity and tracer concentration respectively. Here (in this zonally symmetric experiment - 2D effectively) we define the mean as a zonal and time average. $\mathbf{R}$ is a second rank diffusivity tensor of size $D^2$, where $D$ is the dimension of the problem. 

$\mathbf{R}$ is a general tensor and capable of capturing both, the diffusive and advective, influences of eddies on tracer transport. If $\mathbf{R}$ is constrained to be symmetric, it can only quantify the diffusive transport, whereas as antisymmetric $\mathbf{R}$ captures only the advective transport. Thus, it seems to me, that $\mathbf{R}$ will be able to capture all the transport processes that influence the mean tracer field, as long as the assumption of eddy fluxes being related to the gradient of tracer field are appropriate. This is a local assumption, fluxes locally are influenced by local gradients. 

There is a very clear expanation of the assumptions that go into calculating $\mathbf{R}$. 
- Locality of fluxes, both in space and time (Time aspect is the study of Georgy). 
- If one is going to use multiple tracers, then there is the assumption that all the tracers are being advected with the same $\mathbf{R}$.

# Tracer gradients

This section can possibly be updated in terms of the way the gradients are calculated on the smoothed data.

```python
# Function to calculate the tracer gradients and return as an array
# only works on xarrays with proper grids.

def tracer_grads(ds, ds_full):
    
    grid = Grid(ds_full, periodic='None')
    
    dy = ds_full.dyC.mean('XC')
    dz = ds_full.drF
    
    trac_dy = grid.interp(grid.diff(ds,'Y', boundary = 'extend')/dy, 'Y',boundary='extend')
    
    trac_dz = - grid.interp(grid.diff(ds,'Z', boundary = 'extend'), 'Z',boundary='extend')/dz
   
    trac_dy= trac_dy.rename('dcdy')
    trac_dz= trac_dz.rename('dcdz')
    
    grads = xr.merge([trac_dy, trac_dz])
    return grads
```

```python
# Calculate tracer gradients (dC/dy, dC/dz)


for k in list_tracers:
    tracGrad = tracer_grads(trac_Tav_Xav[k], ds)
    
    trac_Tav_Xav[k+'_Y'] = tracGrad['dcdy']
    trac_Tav_Xav[k+'_Z'] = tracGrad['dcdz']
```

# Make function that will do the inversion to find diffusivity tensor

```python
# Function to calculate the diffusivity tensor 

def calc_tensor_2D(vc,wc, cy,cz):
    Aflux = np.array([vc, wc])
    Agrad = np.array([cy, cz])


    if ~(np.isnan(Agrad).any() | np.isnan(Aflux).any()):
        return -(Aflux.dot(pinv(Agrad)))
    else:
        return np.nan*(Aflux.dot(Agrad.T))
```

# Spatial averaged 


The output is time averaged variables (tracers and velocities) ($\overline{A}$) and fluxes ($\overline{AB}$).

This mean that the transient eddy fluxes can be calculated as 
$$ 
\overline{A'B'} = \overline{AB} - \overline{A}\overline{B}
$$

Now we want to add spatial averaging ($<.>$). This should involve in the definition of new eddy fluxes, simply by using the relation above.

```python
def sort_ev(x, xvec):
    xreal = x.real
    y = np.abs(xreal)
    if not np.isnan(y).any():
        z = y.argsort(axis=-1)
        return xreal[z.data], xvec[:,z.data]
    else:
        return xreal, xvec
```

```python
# split into symmetric and asymmetric parts 

Ktensor_stxavT = Ktensor_stxav.transpose('Yicoarse','month','Z','j','i') 

ST = 0.5*(Ktensor_stxav.data + Ktensor_stxavT.data)
AT = 0.5*(Ktensor_stxav.data - Ktensor_stxavT.data)

ST = xr.DataArray(ST, coords=Ktensor_stxav.coords, dims=Ktensor_stxav.dims)
AT = xr.DataArray(AT, coords=Ktensor_stxav.coords, dims=Ktensor_stxav.dims)

# calculate the eigenvalues and eigenvectors of symmetric part
eigvalsST, eigvecsST = xr.apply_ufunc(eig, ST, input_core_dims=[['i','j']],
                                    vectorize=True, output_core_dims=[['ii'], ['k','ii']])

   
# the eigenvectors have been arranged based on their magnitudes, while we want to arrange based on absolute magnitude
# so we need to resort 
evalsST, evecsST = xr.apply_ufunc(sort_ev, eigvalsST, eigvecsST, input_core_dims=[['ii'],['k','ii']]
                       , output_core_dims=[['ii'],['k','ii']], vectorize=True)
# now the eigenvalues are arranged from smallest to largest. 

evalsST = evalsST.transpose('month', 'Z', 'Yicoarse','ii')
```

```python
# split into symmetric and asymmetric parts 

Ktensor_stxav_meanT = Ktensor_stxav_mean.transpose('Yicoarse','Z','j','i') 

ST_mean = 0.5*(Ktensor_stxav_mean.data + Ktensor_stxav_meanT.data)
AT_mean = 0.5*(Ktensor_stxav_mean.data - Ktensor_stxav_meanT.data)

ST_mean = xr.DataArray(ST_mean, coords=Ktensor_stxav_mean.coords, dims=Ktensor_stxav_mean.dims)
AT_mean = xr.DataArray(AT_mean, coords=Ktensor_stxav_mean.coords, dims=Ktensor_stxav_mean.dims)

# calculate the eigenvalues and eigenvectors of symmetric part
eigvalsST_mean, eigvecsST_mean = xr.apply_ufunc(eig, ST_mean, input_core_dims=[['i','j']],
                                    vectorize=True, output_core_dims=[['ii'], ['k','ii']])

    
# the eigenvectors have been arranged based on their magnitudes, while we want to arrange based on absolute magnitude
# so we need to resort 
evalsST_mean, evecsST_mean = xr.apply_ufunc(sort_ev, eigvalsST_mean, eigvecsST_mean, input_core_dims=[['ii'],['k','ii']]
                       , output_core_dims=[['ii'],['k','ii']], vectorize=True)
# now the eigenvalues are arranged from smallest to largest. 

evalsST_mean = evalsST_mean.transpose('Z', 'Yicoarse','ii')
```

# What have we done so far? 

At this point we stand with the complete diffusivity tensor $\mathbf{R}$, which has been calculated using the multiple tracer method. As we are operating in a zonally symmetric (and zonally averaged domain), this is a function of only 2 coordinates (Y,Z). 

The symmetric part of this tensor tells us about the diffusive component of the tracer fluxes.  
The antisymmetric part of this tensor, which can be written as a stream function (scalar in 2D), tells us about the advective component of the tracer fluxes. 

An antisymmetric tensor ($\mathbf{A}$) was calculated, which is the stream function of the advective part of the transport.  
As a symmetric tensor multiplied (inner product) with a antisymmetric tensor is always zero, the vector (with only 1 component in 2D) imbedded in the antisymmetric part is guaranted to be divegenceless. 

However - this antisymmetric tensor ($\mathbf{A}$) implies a flux of tracer (buoyancy included), which is $F = -\mathbf{A} \nabla C$. This flux is skew ($F.\nabla C = 0$) by mathmematical identity. This also implies that there is **no rotational part in this flux**, because the rotational part will not be perpendicular to the tracer gradients. We do not need to check this, because it is true for any antisymmetric tensor. 

In 2D over here :  

\begin{equation}
\mathbf{A} = 
\left[\begin{array}{cc}
 0 & -\chi \\
 \chi & 0 \\
\end{array} \right]
\end{equation}

This is equivalent to writing an advective stream function $\psi_A = (-\chi, 0, 0)$,  
which gives an advective velocity of the form $\mathbf{v}_A = (0, -\chi_z, \chi_y)$.  
The equivalent flux is $\mathbf{F_A} = \mathbf{v}_A C = - \mathbf{A} \nabla C$.

# What tests can be done with where we have reached? (obviously to know if we are on the correct path).
In 2D we know things workout, and just qualitative checks are "good enough". 

 - Error estimates - Quantitative. We can calculate an error estimate of the form (what fraction of the tracer flux can be described by the diagnosed $\mathbf{R}$) : 
 \begin{equation}
 E (\overline{u_i' c'}) = \frac{|\overline{u_i'c'} - R_{ij} \partial_jC | }{\overline{u_i'c'}}.
 \end{equation}
 This can obviously be done for each individual tracer, or even for the buoyancy if we believe that the same $\mathbf{R}$ can explain the buoyancy fluxes. 
 - Qualitative - We can also look at the direction of the eigenvectors of the diffusivity, to see if it is aligned with the isopycnal slopes. 
 - Qualitative - Does the structure of S,A, R etc look reasonable? Mid-depth maxima, elevated values near the center of the domain and in correspondence to EKE? 
 - How is the $\chi$ (the stream function corresponding to $\mathbf{A}$ related to the eddy driven overturning circulation ? (Figure 3 - Abernathey et al 2013). 
 
 

# Issues : Gauge and Rotational fluxes (They are different!) 

Rotational fluxes are dropped out, in this zonal average, because according to Marshall and Shutts 1981 the integral of rotational fluxes following a closed contour of tracer variance is 0. 

What about the gauge? 
- This is an arbitrary potential that can be added to a stream function, because it does not change the corresponding velocity field. This leads to the non-uniqueness of the stream function - unless boundary conditions on the stream function are imposed.

# Buoyancy diffusion (GM)
 
This is a **particular** parameterization with 2 of the following properties:  
a) Does not diffuse buoyancy across its gradient $\mathbf{F}_{GM} . \nabla b =0$.  
b) Amount of APE should be reduced. This is achieved by assuming a linear relationship between the transfer coefficient and slope of isopycnals (* Is there a reason for this? Could be propotional to slope^2 or anything, right? *). 

The parameterization is in terms of a streamfunction of the form (writing in 2D): 
\begin{equation}
v_{GM} = -(\kappa_{GM}S)_z, w_{GM} = (\kappa_{GM}S)_y 
\end{equation}
where $S = - b_y/b_z$ is the slope of the isopycnals. So $\psi_{GM} = (\kappa_{GM} S, 0, 0)$. 

## How is this related to $\mathbf{A}$? 
We know that $\mathbf{F}_{GM} = v_{GM}b$, it is an advective flux of mean buoyancy (b).  
$\mathbf{F}_{GM} . \nabla b = 0$.  

Also, let $\mathbf{F}_A = - \mathbf{A} \nabla b$ - consider the buoyancy diffusion by the antisymmetric diffusion tensor. 

*Is ${F}_A = {F}_{GM}$? *

It is true that $\nabla.{F}_A = \nabla.{F}_{GM} = \nabla.(\mathbf{v}_{ad} b)$, because neither of these divergences change the domain integrated variance of the buoyancy (following from Vallis book). *Can anything more be said?*

We know that $\mathbf{v}_{ad} = \nabla \times \mathbf{\psi}_{GM} = \nabla \times \mathbf{\psi}_A$. The velocities corresponding to the **two** different stream functions are the same. The stream functions can be different by a gauge - a potential that has curl zero. 

In zonally averaged 2D system an arbitrary gauge vector has the form $(0, \gamma_y, \gamma_z)$, because of zonal uniformity. This makes 2D *very special*, because now even thought the stream function vectors corresponding to GM and $\mathbf{A}$ are not the same, the zonal components - the only component that we care about - are exactly equivalent. This means if we can measure $\mathbf{A}$ then we know $\kappa_{GM}S$ directly, and knowing $S$ then tells us the transfer coefficient.  

Abernathey et al 2013 used this fact about the gauge, in an unsaid manner. 


## Diagnosing GM diffusivity :
As explained above knowing $\mathbf{A}$ and the slope $S$ gives us the exact $\kappa_{GM}$.
There is an alternate way to measure this, as long as the system is adiabatic.

GM parameterization is a parameterization for the adiabatic component of the eddy buoyancy fluxes. Additionally, it is a pameterization for the *skew* part of the flux.  

In general the advective flux (corresponding to an antisymmetric diffusion matrix) has a rotational and divergent part. The divergence of the rotational component is zero, and the divergent component is skew (perpendicular to the tracer gradients). For flux of buoyancy skew means along mean isopycnal eddy flux of isopycnal variability ($\overline{\mathbf{v}'b'}_{sk}.\nabla b = 0$. *Note* how I have carefully labeled the eddy flux as only the skew component, this is because there can be a rotational component and I am not sure what that does in relation to the mean tracer gradients. All, I know about the rotational component is that its divergence is zero, so it does not influence the tracer equation (or the mean tracer variance ($C^2$) integrated over the domain for that matter - However the rotational part does influence the eddy tracer variance ($\overline{c'^2}$)).

However, (apparently because I have not checked this yet) the rotational tracer flux when integrated along a line of mean tracer variance is zero. So, in our zonally averaged and symmetric system the rotational flux simply drops away.

*Thus using adiabatic and zonal symmetry we can also diagnose the GM diffusivity as ($\overline{v'b'} \sim \overline{v'b'}_{adiabatic,skew}$ in zonal channel simulation below the influence of KPP)*

\begin{equation}
\kappa_{GM} = -\overline{v'b'} / \overline{b_y}
\end{equation}

If flow is indeed **adiabatic** flow $ \overline{\mathbf{u}'b'}.\nabla b = 0 $ - buoyancy fluxes are perpendicular to the bupyancy surfaces. *We can check this!* This would give a second equation to calculate $K_{GM}$. 

\begin{equation}
\kappa_{GM} = \overline{w'b'}/(s^2b_z)
\end{equation}

Both these relationships give "approximately" correct results because the model is not perfectly adiabatic.
