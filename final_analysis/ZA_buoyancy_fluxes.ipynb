## Buoyancy Fluxes 

In the Z-coordinate models the buoyancy equation has eddies parameterized using the Gent-McWilliams parameterization. This parameterization is known to be very important for the solution that emerges. 

So here we study how the diagnosed tensor does at reconstructing these buoyancy fluxes. 

```python
# load modules 
import xarray as xr
from matplotlib import pyplot as plt
from xgcm import Grid, generate_grid_ds
import numpy as np
%matplotlib inline
```

```python
Model_Tav = xr.open_dataset('./analysis_data_files/ZA_model_Tav.nc')
```

```python
grid = Grid(Model_Tav)
```

```python
ds_T_eddy = xr.Dataset()
```

```python
ds_T_eddy['UpTp'] = (grid.interp(Model_Tav.UTtave, 'X') - 
                     grid.interp(Model_Tav.uVeltave, 'X') * Model_Tav.Ttave)
ds_T_eddy['VpTp'] = (grid.interp(Model_Tav.VTtave, 'Y', boundary='extend') - 
                     grid.interp(Model_Tav.vVeltave, 'Y', boundary='extend') * Model_Tav.Ttave)
ds_T_eddy['WpTp'] = (grid.interp(Model_Tav.WTtave, 'Z', boundary='extend') - 
                     grid.interp(Model_Tav.wVeltave, 'Z', boundary='extend') * Model_Tav.Ttave)
```

```python
dx = 5e3
ds_T_eddy['dTdx'] = grid.interp(grid.diff(Model_Tav.Ttave,'X')/dx, 'X')
ds_T_eddy['dTdy'] = grid.interp(grid.diff(Model_Tav.Ttave,'Y', boundary='extend')/dx, 
                                'Y', boundary='extend')
ds_T_eddy['dTdz'] = grid.interp(grid.diff(Model_Tav.Ttave,'Z', boundary='extend')/
                                grid.diff(Model_Tav.Z, 'Z', boundary='extend'), 
                                'Z', boundary='extend')
```

```python
ds_T_coarse = ds_T_eddy.coarsen(XC=10, YC=10).mean()
```

```python
diff_tensor = xr.open_dataset('./analysis_data_files/diff_tensor_ZA_50km.nc')
```

```python
def flux_reconstruct(tensor, flux, grads):

    recUflux = -(tensor.sel(i=0, j=0)*grads.dTdx + 
                 tensor.sel(i=0, j=1)*grads.dTdy + tensor.sel(i=0, j=2)*grads.dTdz)
    recVflux = -(tensor.sel(i=1, j=0)*grads.dTdx + 
                 tensor.sel(i=1, j=1)*grads.dTdy + tensor.sel(i=1, j=2)*grads.dTdz)
    recWflux = -(tensor.sel(i=2, j=0)*grads.dTdx + 
                 tensor.sel(i=2, j=1)*grads.dTdy + tensor.sel(i=2, j=2)*grads.dTdz)

    errU = np.abs(flux.UpTp - recUflux)/np.abs(flux.UpTp)
    errV = np.abs(flux.VpTp - recVflux)/np.abs(flux.VpTp)
    errW = np.abs(flux.WpTp - recWflux)/np.abs(flux.WpTp)
    
    flux_rec=xr.Dataset({'UpTp':recUflux, 'VpTp':recVflux, 'WpTp':recWflux,
                                'errU':errU, 'errV':errV, 'errW':errW})
    
    return flux_rec
```

```python
rec_flux_T = flux_reconstruct(diff_tensor.Kcorr, ds_T_coarse, ds_T_coarse)
rec_Sflux_T = flux_reconstruct(diff_tensor.STcorr, ds_T_coarse, ds_T_coarse)
rec_Aflux_T = flux_reconstruct(diff_tensor.ATcorr, ds_T_coarse, ds_T_coarse)
```

### Correlations analysis

```python
def flux_rec_corrs_ZA_vs_depth(flux, rec_flux): 
    corrU_pearson = ((flux.UpTp-flux.UpTp.mean(['XC','YC'])) * (rec_flux.UpTp-rec_flux.UpTp.mean(['XC','YC']))
                    ).mean(['XC','YC']) / (((flux.UpTp-flux.UpTp.mean(['XC','YC']))**2).mean(['XC','YC']) 
                                * ((rec_flux.UpTp-rec_flux.UpTp.mean(['XC','YC']))**2).mean(['XC','YC']))**0.5
    
    corrV_pearson = ((flux.VpTp-flux.VpTp.mean(['XC','YC'])) * (rec_flux.VpTp-rec_flux.VpTp.mean(['XC','YC']))
                    ).mean(['XC','YC']) / (((flux.VpTp-flux.VpTp.mean(['XC','YC']))**2).mean(['XC','YC']) 
                                           * ((rec_flux.VpTp-rec_flux.VpTp.mean(['XC','YC']))**2).mean(['XC','YC']))**0.5
    
    corrW_pearson = ((flux.WpTp-flux.WpTp.mean(['XC','YC'])) * (rec_flux.WpTp-rec_flux.WpTp.mean(['XC','YC']))
                    ).mean(['XC','YC']) / (((flux.WpTp-flux.WpTp.mean(['XC','YC']))**2).mean(['XC','YC']) 
                                           * ((rec_flux.WpTp-rec_flux.WpTp.mean(['XC','YC']))**2).mean(['XC','YC']))**0.5
    
    
    corrs = xr.Dataset({'corrU_pearson':corrU_pearson, 'corrV_pearson':corrV_pearson, 'corrW_pearson':corrW_pearson})
    
    return corrs
```

```python
corr_K_UpSt = flux_rec_corrs_ZA_vs_depth(ds_T_coarse.sel(XC=slice(0,800e3), YC=slice(250e3, 1750e3)), 
                                    rec_flux_T.sel(XC=slice(0,800e3), YC=slice(250e3, 1750e3)))

corr_K_DoSt = flux_rec_corrs_ZA_vs_depth(ds_T_coarse.sel(XC=slice(1200e3,2000e3), YC=slice(250e3, 1750e3)), 
                                    rec_flux_T.sel(XC=slice(1200e3,2000e3), YC=slice(250e3, 1750e3)))


corr_S_UpSt = flux_rec_corrs_ZA_vs_depth(ds_T_coarse.sel(XC=slice(0,800e3), YC=slice(250e3, 1750e3)), 
                                    rec_Sflux_T.sel(XC=slice(0,800e3), YC=slice(250e3, 1750e3)))

corr_S_DoSt = flux_rec_corrs_ZA_vs_depth(ds_T_coarse.sel(XC=slice(1200e3,2000e3), YC=slice(250e3, 1750e3)), 
                                    rec_Sflux_T.sel(XC=slice(1200e3,2000e3), YC=slice(250e3, 1750e3)))


corr_A_UpSt = flux_rec_corrs_ZA_vs_depth(ds_T_coarse.sel(XC=slice(0,800e3), YC=slice(250e3, 1750e3)), 
                                    rec_Aflux_T.sel(XC=slice(0,800e3), YC=slice(250e3, 1750e3)))

corr_A_DoSt = flux_rec_corrs_ZA_vs_depth(ds_T_coarse.sel(XC=slice(1200e3,2000e3), YC=slice(250e3, 1750e3)), 
                                    rec_Aflux_T.sel(XC=slice(1200e3,2000e3), YC=slice(250e3, 1750e3)))
```

```python
corr_AS_UpSt = flux_rec_corrs_ZA_vs_depth(rec_Sflux_T.sel(XC=slice(0,800e3), YC=slice(250e3, 1750e3)), 
                                    rec_Aflux_T.sel(XC=slice(0,800e3), YC=slice(250e3, 1750e3)))

corr_AS_DoSt = flux_rec_corrs_ZA_vs_depth(rec_Sflux_T.sel(XC=slice(1200e3,2000e3), YC=slice(250e3, 1750e3)), 
                                    rec_Aflux_T.sel(XC=slice(1200e3,2000e3), YC=slice(250e3, 1750e3)))
```
