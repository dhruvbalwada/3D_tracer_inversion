# Eddy Driven Velocity 

When using a mixing length argument and relating the first derivative of the large scale gradient to the eddy flux, a part of the transport tensor is skew. The effect of this skew transport tensor on the tracers can be represented as an advective flux, upto a rotational flux. 

* 
\begin{equation}
 \tilde{\mathbf{v}} C = C \nabla \times \mathbf{\psi} = \nabla \times (C \mathbf{\psi}) - \nabla C \times \mathbf{\psi}
\end{equation}

It is entirely possible that the skew flux (which is estimated using our MMT method), which is the second term on the RHS might have a non-divergent component embeded in it. However, we do not know what to say about this "hidden" rotational part. 
If an equation, similar to the variance equation, was present where the skew flux could explicitly be related to the properties of the turbulence, we might be able to have some insight.  

Regardless, since we know $\mathbf{\psi}$ we can estimate the eddy driven velocity. 

```python
# load modules 

import numpy as np
import xarray as xr
from xmitgcm import open_mdsdataset
from matplotlib import pyplot as plt
from matplotlib import colors, ticker, cm
from matplotlib.colors import LogNorm
%matplotlib inline
import os 
from xgcm import Grid
from xgcm import generate_grid_ds
```

```python
Lcoarse = 100e3
```

```python
Model_Tav = xr.open_dataset('./outputs_ncfiles/Model_Tav.nc')
```

```python
diff_tensor = xr.open_dataset('./outputs_ncfiles/diff_tensor_'+str(int(Lcoarse//1e3))+'km.nc')
Trac_coarse = xr.open_dataset('./outputs_ncfiles/Trac_coarse_'+str(int(Lcoarse//1e3))+'km')
Vels_coarse = xr.open_dataset('./outputs_ncfiles/Vels_coarse_'+str(int(Lcoarse//1e3))+'km')
T_coarse    = xr.open_dataset('./outputs_ncfiles/T_coarse_'+str(int(Lcoarse//1e3))+'km')
```

```python
grid_ds = xr.Dataset()
grid_ds['TRAC01'] = Trac_coarse['TRAC01']
grid_ds['wVeltave'] = Model_Tav['wVeltave']
```

```python
grid_ds = grid_ds.drop(['wVeltave','XC','YC']).isel(XC=0, YC=0)
```

```python
psi1 = diff_tensor.AT.isel(i=1, j=2).rename('psi1')
psi2 = -diff_tensor.AT.isel(i=0, j=2).rename('psi2')
psi3 = diff_tensor.AT.isel(i=0, j=1).rename('psi3')
```

$$ \tilde{\mathbf{v}} = \nabla \times \psi$$

$$\tilde{\mathbf{v}} = \mathbf{i} (\psi_{3y} - \psi_{2z}) + \mathbf{j} (\psi_{1z} - \psi_{3x}) +\mathbf{k} (\psi_{2x} - \psi_{1y}) $$

```python
# Sections of the stream functions 

plt.figure(figsize=(12,3))
plt.subplot(131)
psi1.sel(XCcoarse=0, method='nearest').plot.contourf(vmin=-1)
plt.title('psi1')

plt.subplot(132)
psi2.sel(XCcoarse=0, method='nearest').plot.contourf(vmin=-1)
```

```python
# calculate gradients on coarse grid 
ds = generate_grid_ds(grid_ds, {'X': 'XCcoarse', 'Y':'YCcoarse'})
```

```python
grid = Grid(ds, periodic='X')
```

```python
grid
```

```python
psi1_Y = grid.interp(grid.diff(psi1, 'Y', boundary='extend')/ Lcoarse, 'Y', boundary='extend')
psi2_X = grid.interp(grid.diff(psi2, 'X')/ Lcoarse, 'X')

psi1_Z = -grid.interp(grid.diff(psi1, 'Z', boundary='extend'), 'Z', boundary='extend')/ grid_ds.drF
psi2_Z = -grid.interp(grid.diff(psi2, 'Z', boundary='extend'), 'Z', boundary='extend')/ grid_ds.drF

psi3_X = grid.interp(grid.diff(psi3, 'X')/ Lcoarse, 'X')
psi3_Y = grid.interp(grid.diff(psi3, 'Y', boundary='extend')/ Lcoarse, 'Y', boundary='extend')
```

```python
Vstar1 = (psi3_Y - psi2_Z)
Vstar2 = (psi1_Z - psi3_X)
Vstar3 = (psi2_X - psi1_Y)
```

```python
sx = -T_coarse.T_X / T_coarse.T_Z
sy = -T_coarse.T_Y / T_coarse.T_Z
s = (sx**2 + sy**2)**0.5
gradb = (T_coarse.T_X**2 + T_coarse.T_Y**2 + T_coarse.T_Z**2)**0.5
```

# Rotate to buoyancy coordinates 

$$\tilde{\psi}_1 = {\psi}_3 \frac{s}{\sqrt{1+s^2}} + {\psi}_1 \frac{s^x}{s\sqrt{1+s^2}}  + {\psi}_2 \frac{s^y}{s\sqrt{1+s^2}} $$


$$\tilde{\psi}_2 = {\psi}_2 \frac{s^x}{s} - {\psi}_1 \frac{s^y}{s} $$

$$\tilde{\psi}_3 = {\psi}_3 \frac{1}{\sqrt{1+s^2}} - {\psi}_1 \frac{s^x}{\sqrt{1+s^2}}  - {\psi}_2 \frac{s^y}{\sqrt{1+s^2}} $$

```python
psi1tilde = 1/(1+s**2)**0.5 *( psi3*s + psi1*sx/s + psi2*sy/s)
psi2tilde = 1/s *(psi2*sx - psi1*sy)
psi3tilde = 1/(1+s**2)**0.5 *( psi3 - psi1*sx  - psi2*sy)
```
