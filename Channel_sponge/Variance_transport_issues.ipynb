```python
flux = xr.Dataset({'UpCp':UpCp, 'VpCp':VpCp, 'WpCp':WpCp})
grads = xr.Dataset({'dCdx':dCdx, 'dCdy':dCdy, 'dCdz':dCdz})
```

```python
def flux_reconstruct(tensor, flux, grads):

    recUflux = -(tensor.sel(i=0, j=0)*grads.dCdx + tensor.sel(i=0, j=1)*grads.dCdy + tensor.sel(i=0, j=2)*grads.dCdz)
    recVflux = -(tensor.sel(i=1, j=0)*grads.dCdx + tensor.sel(i=1, j=1)*grads.dCdy + tensor.sel(i=1, j=2)*grads.dCdz)
    recWflux = -(tensor.sel(i=2, j=0)*grads.dCdx + tensor.sel(i=2, j=1)*grads.dCdy + tensor.sel(i=2, j=2)*grads.dCdz)

    errU = np.abs(flux.UpCp - recUflux)/np.abs(flux.UpCp)
    errV = np.abs(flux.VpCp - recVflux)/np.abs(flux.VpCp)
    errW = np.abs(flux.WpCp - recWflux)/np.abs(flux.WpCp)
    
    flux_rec=xr.Dataset({'UpCp':recUflux, 'VpCp':recVflux, 'WpCp':recWflux,
                                'errU':errU, 'errV':errV, 'errW':errW})
    
    return flux_rec
```

```python
frb = flux_reconstruct(diff_tensor.Kcorr, fluxT, gradT)
flux_rec_buoy_AT = flux_reconstruct(diff_tensor.ATcorr, fluxT, gradT) 
flux_rec_buoy_ST = flux_reconstruct(diff_tensor.STcorr, fluxT, gradT) 
```

## Variance transport term

```python
TpTp = (Model_Tav.TTtave - Model_Tav.Ttave**2)/2
```

```python
uTv = Model_Tav.uVeltave * grid.interp(TpTp, 'X')
vTv = Model_Tav.vVeltave * grid.interp(TpTp, 'Y', boundary='extend')
wTv = Model_Tav.wVeltave * grid.interp(TpTp, 'Z', boundary='extend')
```

```python
L=5e3
```

```python
divvT = (grid.diff(uTv, 'X')/L + grid.diff(vTv, 'Y', boundary='extend')/L 
        - grid.diff(wTv, 'Z', boundary='extend')/Model_Tav.drF)
```

```python
idmax   = int( (Model_Tav.dyC[0,0].values * len(Model_Tav.YC))/Lcoarse) # numper of grid points to average over
nreps   = int( len(Model_Tav.YC)/idmax )
```

```python
# Coarse grain velocities 

YCicoarse = np.repeat(range(idmax),nreps)
XCicoarse = np.repeat(range(idmax),nreps)

# Interpolate mean velocities to all be on same point.# Interp 

divvT.coords['XCicoarse'] = ('XC', XCicoarse)
divvT.coords['YCicoarse'] = ('YC', YCicoarse)


YCcoarse = ('YCicoarse', divvT.YC.groupby('YCicoarse').mean('YC'))
XCcoarse = ('XCicoarse', divvT.XC.groupby('XCicoarse').mean('XC'))

# coarse grain mean velocities


temp = (divvT.groupby('XCicoarse').mean('XC')).groupby('YCicoarse').mean('YC')
    
temp_array = xr.DataArray(temp.data, coords = [temp.Z, YCcoarse, XCcoarse], 
                             dims = ['Z', 'YCcoarse', 'XCcoarse'])
    
divvT_coarse = temp_array
```

```python
FSgb = flux_rec_buoy_ST.UpCp*gradT.dCdx + flux_rec_buoy_ST.VpCp*gradT.dCdy + flux_rec_buoy_ST.WpCp*gradT.dCdz
FAgb = flux_rec_buoy_AT.UpCp*gradT.dCdx + flux_rec_buoy_AT.VpCp*gradT.dCdy + flux_rec_buoy_AT.WpCp*gradT.dCdz
```

```python
def pearson_corr(ds1, ds2):
    corr = ((ds1 - ds1.mean())*(ds2 - ds2.mean())).mean() / ( ((ds1 - ds1.mean())**2).mean()**0.5 * ((ds2 - ds2.mean())**2).mean()**0.5 )
    return corr 
```

```python
corrUS = pearson_corr(divvT_coarse.sel(YCcoarse=slice(300e3, 1700e3), Z=slice(-300,-2000), XCcoarse=slice(0,800e3)), 
                    -FSgb.sel(YCcoarse=slice(300e3, 1700e3), Z=slice(-300,-2000), XCcoarse=slice(0,800e3)))


corrDS = pearson_corr(divvT_coarse.sel(YCcoarse=slice(300e3, 1700e3), Z=slice(-300,-2000), XCcoarse=slice(1200e3,2000e3)), 
                    -FSgb.sel(YCcoarse=slice(300e3, 1700e3), Z=slice(-300,-2000), XCcoarse=slice(1200e3,2000e3)))
corrUS, corrDS
```
